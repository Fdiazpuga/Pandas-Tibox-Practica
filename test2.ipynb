```python
from openpyxl import load_workbook
import pandas as pd

from datetime import date
from datetime import datetime

wb = load_workbook(filename='C:\\Users\\Fefi\\Desktop\\2022 Carga MBI v3.xlsx', data_only=True)
ws = wb['MBI META']

data = ws.values

columns1 = next(data)[0:]
columns2 = next(data)[1:]
columns3 = next(data)[2:]
columns4 = next(data)[3:]
columns5 = next(data)[0:28] # Encabezado fila 5

data = list(data)

df = pd.DataFrame([x if len(x) == 28 else x[:28] for x in data], columns=columns5) # Si hay mas columnas las ignora (ej: columna insert al final)
df = df.dropna() # Remover filas vacias (por algun motivo toma como 2mil filas vacias del final)

print (df.dtypes)
#df.head(30)
display(df)
```

```python
from datetime import date
from datetime import datetime

# Funcion para transformar las columnas que vienen con formato fecha a int
# a: nombre columna
# b: numero que se le suma porque por algun motivo le faltan (puede que sea por aproximacion)
def fu(a,b):
    for i in df.index:
        if len(str(df.at[i, a])) != None:
            try:
                f_date = date(1900,1,1)
                l_date = (df.at[i, a].date())
                delta = l_date - f_date
                df.at[i, a] = (delta.days+2) # +b
            except AttributeError:
                df.at[i, a] = '-'

```
